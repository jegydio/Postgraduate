# Definições



Análise exploratória de dados (AED) ou *EDA - Exploratory Data Analysis* - é uma abordagem de análise de conjuntos de dados, geralmente com o auxílio de técnicas de visualização de dados, para entender suas principais características. A ideia é explorar ao máximo as informações disponíveis antes de formular hipóteses, criar modelos ou fazer inferências. A AED é fundamental na ciência de dados e na análise estatística.

A AED pode incluir o seguinte:

* **Resumo estatístico:** O primeiro passo na AED é geralmente obter um resumo estatístico dos dados. Isso inclui medidas de centralidade como média, mediana e moda, e medidas de dispersão como variação e desvio padrão.

* **Análise de formas de distribuição:** Este é um passo importante para entender como os dados estão distribuídos. Isso pode incluir a construção de histogramas, diagramas de caixa (boxplots) e gráficos de densidade.

* **Análise de relação entre variáveis:** Se houver múltiplas variáveis no conjunto de dados, pode ser útil entender a relação entre elas. Isso pode incluir a construção de gráficos de dispersão e calcular coeficientes de correlação.

* **Identificação de valores atípicos:** Valores atípicos podem ter um grande efeito na análise, portanto, é importante identificá-los durante a fase de AED.

* **Testes de hipóteses:** Embora a AED seja principalmente sobre a exploração sem qualquer hipótese prévia, às vezes os insights obtidos durante a AED podem levar a testes de hipóteses para confirmar se uma observação é estatisticamente significativa.

A análise exploratória de dados é um componente essencial de qualquer análise de dados e ajuda a entender a estrutura subjacente dos dados, identificar outliers ou anomalias, e encontrar padrões interessantes nos dados.

As funções `pd.to_numeric()` e `pd.to_datetime()` são funções da biblioteca pandas em Python para converter os dados de uma coluna específica em tipos numéricos e de data/hora, respectivamente. Aqui está uma explicação de cada uma delas:

1. `pd.to_numeric(df[a]):`

* `df[a]`: É a coluna do DataFrame df que você deseja converter para um tipo numérico.
* `pd.to_numeric()`: É uma função que converte os valores da coluna em números. Se houver algum valor inválido ou não numérico na coluna, o pandas converterá esse valor para `NaN`.

* Sintaxe correta: `df[a] = pd.to_numeric(df[a])`

2. `pd.to_datetime(df[a])`:

* `df[a]`: É a coluna do DataFrame df que você deseja converter para o tipo de data/hora.
* `pd.to_datetime()`: É uma função que converte os valores da coluna em objetos de data/hora do pandas.
* Sintaxe correta: `df[a] = pd.to_datetime(df[a])`

```python
# Seaborn vem com estilos predefinidos que melhoram a aparência dos gráficos
sns.set_theme()

# Criar figura e eixos
fig, ax1 = plt.subplots(figsize=(18, 6))

# Plotar a primeira coluna com estilo do seaborn
sns.lineplot(data=df_total_br, x=df_total_br.index, y='casosAcumulado', color='blue', ax=ax1)
ax1.set_xlabel('Data')
ax1.set_ylabel('Casos Acumulados', color='blue')

# Adicionar grade ao gráfico
ax1.grid(True)

# Criar o segundo eixo y
ax2 = ax1.twinx()

# Plotar a segunda coluna com estilo do seaborn
sns.lineplot(data=df_total_br, x=df_total_br.index, y='obitosAcumulado', color='red', ax=ax2)
ax2.set_ylabel('Óbitos Acumulados', color='red')

# Definir título
plt.title('Gráfico Óbitos e Casos Acumulados')

# Formatar data no eixo x
ax1.xaxis.set_major_locator(mdates.MonthLocator())
ax1.xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))

# Selecionar cada segunda marca
xticks = ax1.get_xticks()
ax1.set_xticks(xticks[::2]) # escolhe cada segunda marca

# Rotacionar rótulos do eixo x
plt.setp(ax1.get_xticklabels(), rotation=45)

# Ajustar o espaçamento entre subplots
fig.tight_layout()

# Exibir o gráfico
plt.show()
```

```python
import requests

# O URL para a API
url = 'https://imunizacao-es.saude.gov.br/_search'

# Faz a requisição para a API
response = requests.get(url, auth=('imunizacao_public', 'qlto5t&7r_@+#Tlstigi'))

# Verifica se a requisição foi bem sucedida
if response.status_code == 200:
    # Converte a resposta em JSON para um dicionário Python
    data = response.json()

    # Supondo que os dados que você quer estão na chave 'hits' do dicionário,
    # e dentro disso, cada 'hit' tem um campo '_source' que contém os dados relevantes
    hits = data.get('hits', {}).get('hits', [])

    # Agora, você pode criar uma lista de todos os campos '_source' e transformá-los em um DataFrame
    source_data = [hit.get('_source', {}) for hit in hits]

    df = pd.DataFrame(source_data)
    print(df)
else:
    print('Erro na requisição:', response.status_code)
```

```python
# O URL para a API
url = 'https://imunizacao-es.saude.gov.br/_search'

# O tamanho da página (número de hits por página)
page_size = 10000

# A página inicial
page = 0

# Uma lista para armazenar todos os dados
data = []

# Loop até que não haja mais dados
while True:
    # A consulta para a API
    query = {
        "size": page_size,
        "from": page * page_size,
        "_source": ["vacina_dataAplicacao", "vacina_descricao_dose"],
        "query": {
            "match_all": {}
        }
    }

    # Faz a requisição para a API
    response = requests.get(url, json=query, auth=('imunizacao_public', 'qlto5t&7r_@+#Tlstigi'))

    # Verifica se a requisição foi bem sucedida
    if response.status_code != 200:
        print('Erro na requisição:', response.status_code)
        break

    # Converte a resposta em JSON para um dicionário Python
    page_data = response.json()

    # Extrai os hits da resposta
    hits = page_data.get('hits', {}).get('hits', [])

    # Se não houver mais hits, sai do loop
    if not hits:
        break

    # Extrai os dados relevantes dos hits e adiciona à lista de dados
    data.extend(hit.get('_source', {}) for hit in hits)

    # Incrementa a página para a próxima iteração
    page += 1

# Converte os dados em um DataFrame
df_vacinacao = pd.DataFrame(data)

print(df_vacinacao)
```

```python
# convertendo a coluna Data para o formato data

df_vacinacao['vacina_dataAplicacao'] = pd.to_datetime(df_vacinacao['vacina_dataAplicacao']).dt.strftime('%Y-%m-%d')
df_vacinacao.set_index('vacina_dataAplicacao', inplace=True)
```

# Web Scraping

Web scraping é uma técnica de extração de dados da web, que consiste em coletar informações de sites de forma automatizada. O termo "scraping" vem do inglês e significa "raspar" ou "coletar raspando".

O processo de web scraping envolve o acesso a uma página da web, a análise da estrutura HTML dessa página e a extração dos dados desejados. Isso é feito utilizando bibliotecas ou frameworks específicos em uma linguagem de programação, como Python.

Os dados podem ser extraídos de várias maneiras, incluindo a análise do HTML da página para identificar elementos específicos, como tags, classes ou IDs, e extrair o conteúdo dentro desses elementos. Além disso, é possível seguir links para outras páginas e repetir o processo de extração em várias páginas relacionadas.

Existem diversas ferramentas e bibliotecas populares em Python para facilitar o web scraping, como BeautifulSoup, Selenium, Scrapy e Requests. Essas bibliotecas fornecem funcionalidades para fazer requisições HTTP, analisar o HTML, extrair dados estruturados e até mesmo interagir com páginas da web que dependem de JavaScript.

No entanto, ao realizar web scraping, é importante conhecer e respeitar as leis e políticas de uso dos sites, bem como evitar sobrecarregar os servidores com um grande número de requisições em curtos intervalos de tempo.
